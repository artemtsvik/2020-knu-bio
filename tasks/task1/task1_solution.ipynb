```python
from bn256 import *
import random
from copy import copy
```

```python
r1 = random.randrange(2, order)
r2 = random.randrange(2, order)
r3 = random.randrange(2, order)
```

```python
r1, r2, r3
```

```python
T1 = curve_G.scalar_mul(r1)
T2 = curve_G.scalar_mul(r2)
T3 = curve_G.scalar_mul(r3)
```

```python
T1, T2, T3
```

```python
def inverse_curve(curve):
    inv = copy(curve)
    inv.y = inv.y.additive_inverse()
    return inv
```

```python
# test
point_add(T1, inverse_curve(T1))
```

```python
# TP = T1 + T2 - T3
TP = point_add(point_add(T1, T2), inverse_curve(T3))
```

```python
TP
```

```python
T4 = twist_G.scalar_mul(r1)
T5 = twist_G.scalar_mul(r2)
T6 = twist_G.scalar_mul(r3)
```

```python
T4, T5, T6
```

```python
def inverse_twist(twist):
    inv = copy(twist)
    inv.negate()
    return inv
```

```python
# test
point_add(T6, inverse_twist(T6))
```

```python
# TQ = T4 + T5 - T6
TQ = point_add(point_add(T4, T5), inverse_twist(T6))
```

```python
TQ
```

```python
tf = r1 + r2 - r3
```
